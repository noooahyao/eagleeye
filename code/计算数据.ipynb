```python
import numpy as np
import pandas as pd
import os
import re
```

```python
read_csv=pd.read_excel("new_feature.xlsx",index_col="属性")
read_csv.head()
```

```python
firm=[]
os.chdir("./数据")
path=os.getcwd()
for i in os.listdir(path):
    if '.' not in i:
        firm.append(i)
```

```python
import shutil
name=firm[0]
os.chdir(os.path.join(path,name))
df=pd.read_excel('%s.xlsx'%name,sheet_name='data')
df.head()
```

```python
feature_name=[]
for i in np.unique(read_csv.index):
    print(i)
    if i in ['硕士','本科','技术']:
        x_per=df[i]/df['员工总人数']
        x_per=x_per.fillna(0)
        x_per=x_per.tolist()
    else:
        x_per=df[i].tolist()
    standard=[j.split(',') for j in read_csv.loc[i]['评分标准'].tolist()]
    number= read_csv.loc[i]['对应分数'].tolist()
    per=read_csv.loc[i]['板块评分占比'].tolist()[0]
    result=[]
    for x in x_per:
        for j in range(len(standard)):
            if x>=float(standard[j][0]) and x<=float(standard[j][1]):
                result.append(number[j]*per)
                break
    df[i+'评分']=result
    feature_name.append(i+'评分')
df=df[feature_name]
df.to_excel("%s指标评分.xlsx"%name)
```
